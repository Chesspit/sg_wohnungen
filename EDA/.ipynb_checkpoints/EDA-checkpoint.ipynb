df = pd.read_csv('belegung-neu-erstellter-wohnungen-2011-2019.csv',  sep=";", parse_dates=['Monat_Jahr'])
df=df.drop(df.columns[0:1], axis=1)

```python
df.info()
```

```python
df.describe()
```

```python
# Combine 'EGID_chiff' and 'EWID' into a new column named 'ID'
df['ID'] = df['EGID_chiff'].astype(str) + '-' + df['EWID'].astype(str)

# Reorder the columns to place 'Building_Apartment_ID' after 'EGID_chiff'
column_order = ['EGID_chiff', 'EWID', 'ID'] + [col for col in df.columns if col not in ['EGID_chiff', 'EWID', 'ID']]
df = df[column_order]

```

```python
df.info()
```

```python
# First, we'll sort the data by 'Building_Apartment_ID' and 'Belegungsjahr' to prepare for change detection.
df_sorted = df.sort_values(by=['ID', 'Belegungsjahr'])

# Create a column to flag changes in household composition
# We will shift the data in the sorted dataframe and compare it with the original to detect changes.

df_sorted['Wechsel'] = (
    (df_sorted['HH_Typ'] != df_sorted['HH_Typ'].shift())
) & (df_sorted['ID'] == df_sorted['ID'].shift()) & (df_sorted['Belegungsjahr'] != df_sorted['Belegungsjahr'].shift()) & (df_sorted['Belegungsjahr'] != df_sorted['Belegungsjahr'].min())

# Since we're interested in changes within the same apartment, we need to ensure that the shift operation
# doesn't mix different apartments, hence the condition on 'ID'.

```

```python
df_sorted.info()
```

```python
# Count the number of unique 'ID' values in the dataframe
unique_id_count = df_sorted['ID'].nunique()
unique_id_count 
```

```python
df_sorted['Dichte'] =  (df_sorted['Anz_Erwachsene'] + 0.5 * df_sorted['Anz_Kinder']) / df_sorted['Wohnungsgroesse']
```

```python
df_sorted['Ueberbelegung'] = (df_sorted['Anz_Erwachsene'] + 0.5 * df_sorted['Anz_Kinder']) >= (df_sorted['Wohnungsgroesse'] + 2)
```

```python
# Create a new column 'Groesse_Matrix' based on the 'Wohnungsgroesse' column
df_sorted['WGM'] = df_sorted['Wohnungsgroesse'].apply(lambda x: '6+' if x >= 6 else x)
```

```python
# Drop the "Quartiergruppe" from the "Quartiergruppe Name" in all entries
df_sorted['Quartiersgruppe Name'] = df_sorted['Quartiersgruppe Name'].str.replace('Quartiergruppe ', '')

```

```python
# Rename a specific string in the column "Quartiersgruppe Name"
df_sorted['Quartiersgruppe Name'] = df_sorted['Quartiersgruppe Name'].replace('Linsebühl - Dreilinden', 'Linsebühl')
```

```python
# Rename a specific string in the column "Quartiersgruppe Name"
df_sorted['Quartiersgruppe Name'] = df_sorted['Quartiersgruppe Name'].replace('Langgass - Heiligkreuz', 'Langgass')
```

```python
df_sorted.info()
```

```python
df_sorted.head()

```

```python
# Calculate the average number of 'wohnungswechsel' per 'Quartiersgruppe Name' and 'Wohnungsgroesse'
avg_wechsel = df_sorted.groupby(['Quartiersgruppe Name', 'Wohnungsgroesse'])['Wechsel'].mean().reset_index()

# Pivot the data to get 'Quartiersgruppe Name' as columns and 'Wohnungsgroesse' as rows
wechsel_matrix = avg_wechsel.pivot_table(index='Wohnungsgroesse', columns='Quartiersgruppe Name', values='Wechsel')

# Display the resulting pivot table
wechsel_matrix
```

```python
df_filtered = df_sorted[(df_sorted['Belegungsjahr'] >= 2013) & (df_sorted['Belegungsjahr'] <= 2017)]
```

```python
import plotly.express as px

# Calculate the average 'Dichte' per 'Quartiergruppe Name' and 'Wohnungsgroesse'
avg_dichte = df_filtered.groupby(['Quartiersgruppe', 'Wohnungsgroesse'])['Dichte'].mean().reset_index()
```

```python
avg_dichte.head
```

```python
dichte_matrix = avg_dichte.pivot_table(index="Wohnungsgroesse", columns="Quartiersgruppe", values="Dichte")
```

```python
df_grouped = df_sorted.groupby(['Quartiersgruppe Name', 'Wohnungsgroesse'])['ID'].nunique().reset_index()
df_pivot = df_grouped.pivot(index='Wohnungsgroesse', columns='Quartiersgruppe Name', values='ID')
```
